```python
# Function to fetch stock data from Alpha Vantage
def fetch_stock_data(symbol, interval):
    """
    Fetch real-time stock data using Alpha Vantage API.

    Args:
        symbol (str): The stock ticker symbol (e.g., "AAPL").
        interval (str): Time interval between data points (e.g., "1min", "5min").
    
    Returns:
        pd.DataFrame: DataFrame containing stock price data.
    """
    base_url = "https://www.alphavantage.co/query"
    params = {
        "function": "TIME_SERIES_INTRADAY",
        "symbol": symbol,
        "interval": interval,
        "apikey": API_KEY,
    }
    response = requests.get(base_url, params=params)
    data = response.json()
    
    # Extract time series data
    if "Time Series" in response.text:
        time_series_key = list(data.keys())[1]  # Identify the key for time series data
        time_series = data[time_series_key]
        
        # Convert to DataFrame
        df = pd.DataFrame(time_series).T
        df.columns = ["Open", "High", "Low", "Close", "Volume"]
        df.index = pd.to_datetime(df.index)  # Convert index to datetime
        df = df.astype(float)  # Convert values to float
        return df
    else:
        print("Error fetching data:", data)
        return None
        
def calculate_moving_averages(data, short_window=5, long_window=20):
    """
    Calculate short-term and long-term moving averages.
    
    Args:
        data (pd.DataFrame): Stock data with a 'Close' column.
        short_window (int): Window size for the short-term moving average.
        long_window (int): Window size for the long-term moving average.
    
    Returns:
        pd.DataFrame: DataFrame with added SMA and LMA columns.
    """
    data['SMA'] = data['Close'].rolling(window=short_window).mean()
    data['LMA'] = data['Close'].rolling(window=long_window).mean()
    return data

def generate_signals(data):
    """
    Generate buy/sell signals based on moving average crossover.
    
    Args:
        data (pd.DataFrame): Stock data with 'SMA' and 'LMA' columns.
    
    Returns:
        pd.DataFrame: DataFrame with 'Signal' column (1=Buy, -1=Sell).
    """
    data['Signal'] = 0  # Initialize with no signal
    data.loc[data['SMA'] > data['LMA'], 'Signal'] = 1  # Buy signal
    data.loc[data['SMA'] < data['LMA'], 'Signal'] = -1  # Sell signal
    return data

def plot_signals(data):
    """
    Plot stock prices with buy/sell signals.
    
    Args:
        data (pd.DataFrame): Stock data with 'Signal' column.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.plot(data['SMA'], label='SMA', color='green', linestyle='--')
    plt.plot(data['LMA'], label='LMA', color='red', linestyle='--')

    # Plot buy signals
    buy_signals = data[data['Signal'] == 1]
    plt.scatter(buy_signals.index, buy_signals['Close'], label='Buy Signal', marker='^', color='green', alpha=1)

    # Plot sell signals
    sell_signals = data[data['Signal'] == -1]
    plt.scatter(sell_signals.index, sell_signals['Close'], label='Sell Signal', marker='v', color='red', alpha=1)

    plt.title("Stock Price with Buy/Sell Signals")
    plt.xlabel("Time")
    plt.ylabel("Price")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def plot_moving_averages(symbol, interval='1Min', limit=100, short_window=5, long_window=20):
    """
    Fetch stock data and plot short-term and long-term moving averages with crossing points.
    
    Args:
        symbol (str): Stock ticker symbol (e.g., 'TSLA').
        interval (str): Time interval (e.g., '1Min', '5Min', etc.).
        limit (int): Number of data points to fetch.
        short_window (int): Window size for the short-term moving average.
        long_window (int): Window size for the long-term moving average.
    """
    # Fetch the last N bars for the given stock
    bars = api.get_bars(symbol, interval, limit=limit)

    # Extract timestamps and closing prices
    timestamps = [bar.t for bar in bars]
    closing_prices = [bar.c for bar in bars]

    # Convert to a pandas DataFrame for easier manipulation
    import pandas as pd
    data = pd.DataFrame({'Timestamp': timestamps, 'Close': closing_prices})
    data['Timestamp'] = pd.to_datetime(data['Timestamp'])
    data.set_index('Timestamp', inplace=True)

    # Calculate moving averages
    data['SMA'] = data['Close'].rolling(window=short_window).mean()
    data['LMA'] = data['Close'].rolling(window=long_window).mean()

    # Identify crossing points
    data['Signal'] = 0
    data.loc[data['SMA'] > data['LMA'], 'Signal'] = 1
    data.loc[data['SMA'] <= data['LMA'], 'Signal'] = -1

    # Plot the data
    plt.figure(figsize=(14, 7))
    plt.plot(data.index, data['Close'], label='Closing Price', color='blue', alpha=0.5)
    plt.plot(data.index, data['SMA'], label=f'{short_window}-period SMA', color='green', linestyle='--')
    plt.plot(data.index, data['LMA'], label=f'{long_window}-period LMA', color='red', linestyle='--')

    # Mark buy signals (crossing upward)
    buy_signals = data[(data['Signal'] == 1) & (data['Signal'].shift(1) == -1)]
    plt.scatter(buy_signals.index, buy_signals['Close'], label='Buy Signal', marker='^', color='green', s=100, alpha=1)

    # Mark sell signals (crossing downward)
    sell_signals = data[(data['Signal'] == -1) & (data['Signal'].shift(1) == 1)]
    plt.scatter(sell_signals.index, sell_signals['Close'], label='Sell Signal', marker='v', color='red', s=100, alpha=1)

    # Add labels and legend
    plt.title(f'{symbol} Price with Short and Long-Term Moving Averages')
    plt.xlabel('Time')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.xticks(rotation=45)
    plt.show()

# Example usage
plot_moving_averages('TSLA', interval='1Min', limit=100, short_window=5, long_window=20)
```

```python
import time
import matplotlib.pyplot as plt
from datetime import datetime

# Initialize lists to store timestamps and prices
timestamps = []
prices = []

# Set up the real-time plot
plt.ion()  # Interactive mode
fig, ax = plt.subplots(figsize=(10, 6))

# Function to fetch the latest TSLA price and update the plot
def update_plot():
    global timestamps, prices
    
    # Fetch the latest TSLA price
    quote = api.get_latest_quote('TSLA')
    latest_price = quote.ask_price  # Use bid_price if preferred

    # Append the new data
    timestamps.append(datetime.now())
    prices.append(latest_price)

    # Clear and update the plot
    ax.clear()
    ax.plot(timestamps, prices, label="TSLA Price", color="blue")
    ax.set_title("Real-Time TSLA Prices")
    ax.set_xlabel("Time")
    ax.set_ylabel("Price (USD)")
    ax.legend()
    ax.grid()

    # Format the x-axis to display time better
    fig.autofmt_xdate()

    # Redraw the plot
    plt.pause(1)

# Run the real-time plot updater
print("Fetching real-time prices. Close the plot window to stop.")
try:
    while True:
        update_plot()
        time.sleep(5)  # Fetch data every 5 seconds
except KeyboardInterrupt:
    print("Stopped fetching real-time prices.")
finally:
    plt.ioff()
    plt.show()
```
